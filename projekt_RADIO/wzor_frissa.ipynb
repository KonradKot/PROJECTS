```python
print('Wykres tłumienia dla wolnej przestrzeni ujetego w sprawozdaniu ,,tłumienie ścian')
import numpy as np
import matplotlib.pyplot as plt
# parametry
f = 21*10**9 #[GHz]
c = 299792458 #[m/s]
lp = 200
d_min = 0
d_max = 2#[cm]

Gr = 1
Gt = 1
# code

# dlug. fali
lamb = c/f
lamb = round(lamb,3)
print(lamb,' [m]')
print(round(lamb*100,3),' [cm]')

zakres = d_max - d_min
d = np.arange(d_min,d_max,zakres/(lp-1),dtype='float')

Pr_Pt = Gr*Gt*((lamb/(4*np.pi*d))**2) #

Pr_Pt = Pr_Pt * 1000

Pr_Pt = 10*np.log10(np.abs(Pr_Pt))

#plot
plt.figure(figsize=(12,8))
plt.title("Wykres tłumienia (ze wzoru Friisa)")
plt.xlabel("odleglosc [m]")
plt.ylabel("wzgledny spadek mocy nadawanej [dBm]")
plt.plot(d,Pr_Pt)
plt.grid()
```
